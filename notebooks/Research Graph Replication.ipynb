## Imports & Settings

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

## Atlas Brasileiro de Desastres Naturais

```python
atlas_bars = [
    ('Estiagem e Seca', 51.32),
    ('Enxurradas', 20.66),
    ('Inundação', 12.04),
    ('Vendavais', 7.07),
    ('Granizo', 4.20),
    ('Movimento de Massa', 1.79),
    ('Alagamento', 1.32),
    ('Erosão', 0.9),
    ('Incêndio', 0.48),
    ('Tornado', 0.12),
    ('Geada', 0.12)
]

df = pd.DataFrame({'disaster': b[0], 'percentage': b[1]} for b in atlas_bars)
```

```python
fig, ax = plt.subplots(figsize=[15, 8])

rectangles = sns.barplot(x=df.index, y='percentage', hue='disaster', data=df, ax=ax, dodge=False)

for p in ax.patches:
    x = p.get_x() + (p.get_width() / 2.)
    y = p.get_height()
    ax.annotate(f"{y}%", (x, y), ha='center', va='bottom', fontsize=15)


ax.set_ylabel('')
ax.set_title('Afetados por desastres naturais (1991-2012)', pad=20, fontsize=18)
plt.legend(fontsize=15)

yrange = range(0, 60, 10)
plt.yticks(yrange, labels=[f'{i}%' for i in yrange], fontsize=15)
plt.xticks(range(df.shape[0]), labels='')

plt.show()
```
